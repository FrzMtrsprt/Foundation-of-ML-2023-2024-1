```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
data = np.loadtxt('dataset/20151026_113')
print(data.shape)
```

```python
fs = 3000  # Frequency
lowcut = 1
highcut = 30
order = 2  # Order of filter
nyq = 0.5*fs  # Nyquist frequency
low = lowcut/nyq
high = highcut/nyq
b, a = signal.butter(order, [low, high], btype='band')  # Butterworth filter

data_filtered = np.zeros(data.shape)
for i in range(data.shape[1]):
    data_filtered[:, i] = signal.lfilter(b, a, data[:, i])  # Filtered data
```

```python
# Two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

ax1.set_title('Raw data')
ax1.plot(data)
ax1.set_ylabel('Amplitude')

ax2.set_title('Filtered data')
ax2.plot(data_filtered)
ax2.set_ylabel('Amplitude')

plt.show()
```
