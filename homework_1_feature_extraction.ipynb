```python
import numpy as np
import pandas as pd
from tsfresh.feature_extraction.extraction import extract_features
from tsfresh.feature_extraction.settings import EfficientFCParameters
```

```python
data_113 = pd.DataFrame(np.loadtxt('dataset/20151026_113_filtered'))
data_114 = pd.DataFrame(np.loadtxt('dataset/20151026_114_filtered'))
data = pd.concat((data_113, data_114))

data.insert(0, 'id', np.arange(len(data)) // 3000)
data.insert(1, 'time', np.arange(len(data)) % 3000)

data
```

```python
parameters = EfficientFCParameters()
features = extract_features(data,
                            default_fc_parameters=parameters,
                            column_id='id', column_sort='time')

features
```

```python
variances = features.var()
variances = variances.sort_values(ascending=False)

variances[0:30]
```
