```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
from tsfresh.feature_extraction.extraction import extract_features
from tsfresh.feature_extraction.settings import EfficientFCParameters
```

```python
data = pd.DataFrame(np.loadtxt('dataset/20151026_113_filtered'))

data.insert(0, 'id', np.arange(len(data)) // 3000)

data
```

```python
parameters = EfficientFCParameters()
features = extract_features(data, default_fc_parameters=parameters, column_id='id')

features
```

```python
selector = VarianceThreshold(threshold=82500)
selector.fit(features)

selected_names = selector.get_support(indices=True)
selected_features = features.iloc[:, selected_names]

selected_features
```
